## Introduccion 

El número de Prandtl es un parámetro adimensional utilizado en mecánica de fluidos y transferencia de calor que relaciona la difusividad térmica y la viscosidad cinemática de un fluido. Este número es importante en el estudio de la transferencia de calor en fluidos, ya que describe la relación entre la capacidad del fluido para transportar calor y su capacidad para disipar la energía mecánica en forma de movimiento. En general, los fluidos con un número de Prandtl alto tienden a ser más eficientes en la transferencia de calor que los fluidos con un número de Prandtl bajo.

### Que relacion tiene el numero de prandtl y la ecuacion de calor 

El número de Prandtl es un número adimensional utilizado para describir el comportamiento de transferencia de calor de un fluido. Está definido como la relación entre la difusividad térmica y la viscosidad cinemática del fluido. El número de Prandtl es importante en la convección natural y forzada, así como en otros procesos de transferencia de calor. En general, los fluidos con un número de Prandtl alto tienden a ser más eficientes en la transferencia de calor que los fluidos con un número de Prandtl bajo.

## Explicacion del problema 

Este es un ejemplo de código para simular la transferencia de calor en un alambre de acero y graficar la evolución de la temperatura a lo largo del tiempo. El código utiliza una aproximación numérica para resolver la ecuación de calor en una malla espacial y en el tiempo.

Se definen las constantes físicas del acero, como su conductividad térmica, densidad y capacidad calorífica, así como el diámetro y longitud del alambre y las condiciones iniciales de temperatura y ambiente. Luego se establecen los parámetros de la simulación, como el tamaño de la malla espacial, el paso temporal y el tiempo máximo de la simulación.

Se crea una malla espacial en la que se almacenarán las temperaturas, y se establecen las condiciones iniciales. Luego, se crea un array para almacenar las temperaturas en el tiempo y se realiza la simulación en el tiempo mediante un bucle for que calcula la temperatura en el tiempo siguiente en cada punto de la malla y almacena las temperaturas en el array.

Finalmente, se grafica la evolución de la temperatura en el alambre a lo largo del tiempo mediante un bucle for que toma muestras de la temperatura en diferentes momentos del tiempo y grafica la temperatura en la posición correspondiente a lo largo del alambre.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las constantes físicas
k = 60 # Conductividad térmica del acero en W/(m*K)
rho = 7850 # Densidad del acero en kg/m^3
cp = 450 # Capacidad calorífica del acero en J/(kg*K)
D = .1 # Diámetro del alambre en metros
L = 1 # Longitud del alambre en metros
T0 = 20 # Temperatura inicial en grados Celsius
Ta = 35 # Temperatura ambiente en grados Celsius

# Definir los parámetros de la simulación
n = 100 # Número de puntos de la malla
dx = L/n # Tamaño de paso espacial
dt = 0.01 # Tamaño de paso temporal
tmax = 3600 # Tiempo máximo de la simulación


x = np.linspace(0, L, n+1)

# Crear un array para almacenar las temperaturas
T = np.zeros((n+1,))


T[:] = T0

# Crear un array para almacenar las temperaturas en el tiempo
T_history = np.zeros((int(tmax/dt)+1, n+1))
T_history[0,:] = T[:]

# Realizar la simulación en el tiempo
for i in range(1, int(tmax/dt)+1):
    for j in range(1, n):
        T[j] += k*dt/(rho*cp*dx*dx) * (T[j+1] - 2*T[j] + T[j-1])
    T[0] = Ta
    T[n] = Ta
    T_history[i,:] = T[:]
    

plt.figure()
for i in range(0, int(tmax/dt)+1, int(tmax/(dt*10))):
    plt.plot(x, T_history[i,:], label='t='+str(i*dt)+' s')
plt.xlabel('Posición (m)')
plt.ylabel('Temperatura (°C)')
plt.title('Evolución de la temperatura en el alambre')
plt.legend()
plt.show()
```

La gráfica muestra la evolución de la temperatura en el alambre a lo largo del tiempo. El eje horizontal representa la posición a lo largo del alambre (en metros), mientras que el eje vertical representa la temperatura (en grados Celsius).

Cada línea de la gráfica corresponde a un momento del tiempo diferente. La leyenda de la gráfica muestra el tiempo en segundos al que corresponde cada línea.

Por lo tanto, se puede leer la gráfica para obtener información sobre cómo cambia la temperatura en diferentes partes del alambre a lo largo del tiempo. Por ejemplo, se puede observar cómo la temperatura en el centro del alambre aumenta más rápidamente que la temperatura en los extremos del alambre, y cómo la temperatura en el alambre en general se acerca a la temperatura ambiente a medida que pasa el tiempo.
